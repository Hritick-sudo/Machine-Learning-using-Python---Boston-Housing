# Noramalize data

Normalize the data with MinMaxScaler

```python
X = df['LSTAT'].values
Y = df['target'].values
#now we have to normalize thecolumns
#the goal of the normalization is to change the values of data such that all lie under a common scale
#in our case it will be zero to one
```

```python
print(Y[:5])
```

```python
#sklearn provides minmax scaler which takes a list of arbitrary values and gives the values in a range 0 to one without affecting the difference
x_scaler = MinMaxScaler()
X = x_scaler.fit_transform(X.reshape(-1,1))
X = X[:, -1]
y_scaler = MinMaxScaler()
Y = y_scaler.fit_transform(Y.reshape(-1,1))
Y = Y[:, -1]
```

```python
print(Y[:5])
```

```python
xtrain , xtest , ytrain , ytest = train_test_split(X, Y, test_size=0.2)
```

```python
def update(m,x,c,t,learning_rate):
    grad_m = sum(2 * x * ((m * x + c) - t))
    grad_c = sum(2 * ((m * x + c) - t))
    m = m - grad_m * learning_rate
    c = c - grad_c * learning_rate
    return m, c
```

```python
def error(m,x,c,t):
    N = x.size
    e = sum(((m * x + c) - t)**2)
    return e * 1 / (2 * N)
```

```python
def gradient_descent(init_m , init_c , x, t, learning_rate, iterations, error_threshold):
    m = init_m
    c = init_c
    error_values = list()
    mc_values = list()
    for i in range(iterations):
        e = error(m, x, c, t)
        if e < error_threshold :
            print("error less than the error threshold thus stopping gradient descent")
            break
        error_values.append(e)
        m, c = update(m,x,c,t,learning_rate)
        mc_values.append((m,c))
    return m, c, error_values, mc_values
```

```python
%%time
init_m = 0.9
init_c = 0
learning_rate = 0.001
iterations = 250
error_threshold = 0.001


m, c, error_values, mc_values = gradient_descent(init_m, init_c, xtrain, ytrain, learning_rate, iterations, error_threshold) 
```

```python
mc_values_anim = mc_values[0:250:5]
```

```python
fig, ax = plt.subplots()
ln, = plt.plot([],[],'-ro',animated = True)


def init():
    plt.scatter(xtest, ytest, color='g')
    ax.set_xlim(0, 1.0)
    ax.set_ylim(0, 1.0)
    return ln,

def update_frame(frame):
    m, c = mc_values_anim[frame]
    x1, y1 = -0.5, m * -.5 + c
    x2, y2 = 1.5, m * 1.5 + c
    ln.set_data([x1,x2], [y1,y2])
    return ln,


anim = FuncAnimation(fig, update_frame, frames = range(len(mc_values_anim)), init_func = init, blit = True)
HTML(anim.to_html5_video())
```
